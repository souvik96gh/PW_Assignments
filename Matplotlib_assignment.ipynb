#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np<br>
np.random.seed(3)<br>
x = 3 + np.random.normal(0, 2, 50)<br>
y = 3 + np.random.normal(0, 2, len(x))<br>

ANSWER:<br>Scatterplot is the plot using which we can visualize how two variables are correlated. Have a visual on the correlation between two variable can be very useful in the further works of feature engineering: https://en.wikipedia.org/wiki/Correlation#/media/File:Correlation_examples2.svg

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

from matplotlib import pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(x=x,y=y,c="#DE7250",alpha=0.7,marker="*")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatterplot for PWSkills Assignment")
plt.grid()
plt.show()
```

import numpy as np<br>
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])<br>
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])<br>

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])<br>

ANSWER:<br>
Subplot is a technique to draw more than one plots in the same window. It is really helpfull to compare and present more than one attribute of a feature. It's widely used in time series analysis

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2= np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3= np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4= np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(12,24))

plt.subplot(411)
plt.plot(x1,y1,color="red")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Plot1")
plt.grid()

plt.subplot(412)
plt.plot(x2,y2,color="green")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Plot2")
plt.grid()

plt.subplot(413)
plt.plot(x3,y3,color="black")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Plot3")
plt.grid()

plt.subplot(414)
plt.plot(x4,y4,color="#A850DE")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Plot4")
plt.grid()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np<br>
company = np.array(["Apple", "Microsoft", "Google", "AMD"])<br>
profit = np.array([3000, 8000, 1000, 10000])<br>

#### ANSWER:<br>
Bar plot or the bar chart is a type of visualization which is mostly used for categorical data with rectangular bars. Each bar represents a category and their heights represent the corresponding value.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(18,5))

plt.subplot(121)
plt.bar(company,profit,width=0.6,color="red")
plt.title("Bar Chart (Usual/vertical)",fontsize=18)
plt.xlabel("Company",fontsize=14)
plt.ylabel("Profit",fontsize=14)
plt.grid()

plt.subplot(122)
plt.barh(company,profit,0.5,color="green")
plt.title("Horizontal Bar Chart",fontsize=18)
plt.xlabel("Profit",fontsize=14)
plt.ylabel("Company", fontsize=14)
plt.grid()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)<br>
box2 = np.random.normal(90, 20, 200)

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(14,8))

plt.subplot(121)
plt.boxplot(box1)
plt.title("Box1",fontsize=18)
plt.ylabel("Distribution1")


plt.subplot(122)
plt.boxplot(box2)
plt.title("Box2",fontsize=18)
plt.ylabel("Dist2")


plt.show()
```
