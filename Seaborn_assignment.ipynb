### Que1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

### ANSWER:
Like matplotlib, seaborn is also a visualization library which is made based on matplotlib. These are among the few functions which is largely used in industry:
#### Scatterplot:
Scatterplot is the plot using which we can visualize how two variables are correlated. Have a visual on the correlation between two variable can be very useful in the further works of feature engineering: https://en.wikipedia.org/wiki/Correlation#/media/File:Correlation_examples2.svg
#### Boxplot: 
It is basically used to show the distribution of a variable/feature. Here the value of median, 1st and 3rd quartile are visible. This plot is basically used to detect outliers.
#### Heatmap: 
It is another plot to see the corelation of all of the features. Helpful in feature selection.
#### Histogram and KDE: 
Histogram plot is used to visualize the distribution of a particular feature and KDE plot helps to estimate the average kernel.
#### Pairplot: 
This giant plot is used see the corelation of all variable in a single plot. Basically, its the scatterplot between all variable. So in the NxN matrix of plots, the number of scatterplot is NP2 and the number of KDE plot is N (N=number of features/variables)


### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
### Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
plt.figure(figsize=(16,4))
sns.lineplot(x="timepoint",y="signal",data=fmri,sizes=(10,8))
plt.show()
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
### Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
plt.figure(figsize=(16,6))
plt.subplot(121)
sns.boxplot(x = "pclass", y = "age", data = titanic)

plt.subplot(122)
sns.boxplot(x = "pclass", y = "fare", data = titanic)
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
plt.figure(figsize=(12,5))
sns.histplot(data=diamonds, x="price", hue="cut")
plt.grid()
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris,hue="species")
```
